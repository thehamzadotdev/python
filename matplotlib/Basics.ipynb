```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x= np.linspace(0,5,11)
x
y= x**2
y
```

```python
plt.plot(x,y)
plt.title("Data in graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

### Subplots in Matplotlib 

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x,x)
plt.subplot(2,2,4)
plt.plot(x**2, y**3)
```

### Object oriented way

```python
fig = plt.figure(figsize=(20,8), dpi=100)
axis1= fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2= fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

### types of plots

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

### saving plot

```python
fig = plt.figure()
axis1= fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
plt.savefig("axis1.png")
```
