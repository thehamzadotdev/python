```python
import numpy as np
import pandas as pd
```

## Finding missing data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df= pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
# for finding any missing coulmns
df.isna().any()
```

## Removing missing values

```python
df.dropna()
```

```python
df.dropna(thresh=3)
```

## Filling null values

```python
df.fillna(0)
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
```

```python
df.fillna(value= values)
```

```python
df.fillna(df.mean())
```
