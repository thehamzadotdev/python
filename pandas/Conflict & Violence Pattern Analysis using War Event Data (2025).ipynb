#### Missing value analysis

```python
# Dropping gwnob because moost values are empty 
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df["start_day"] = pd.to_datetime(df["date_start"], errors='coerce' )
df["end_day"] = pd.to_datetime(df["date_end"], errors='coerce' )
```

### same day events

```python
sameday_events= df[df["start_day"]== df["end_day"]]
sameday_events.count
```

### multi day events

```python
multiday_events= df[df["end_day"]> df["start_day"]]
multiday_events.count
```

### Total deaths in a war 

```python
df["Total_deaths"]= df["deaths_a"]+ df["deaths_b"]+ df["deaths_civilians"]+ df["deaths_unknown"]
df["Total_deaths"]
```

### minimum deaths

```python
df["Total_deaths"].min()
```

### maximum deaths

```python
df["Total_deaths"].max()
```

### Avg deaths

```python
df["Total_deaths"].mean()
```

### 🔹 SECTION 2: Conflict & Violence Analysis

```python
df.head()
```

### Number of events according to type of voilence

```python
event_per_type = df.groupby("type_of_violence").size()
print(event_per_type)
```

### Average deaths per event

```python
avg_death_per_event= df.groupby("type_of_violence")["Total_deaths"].mean()
avg_death_per_event
```

### Most lethal

```python
most_lethat = avg_death_per_event.sort_values(ascending = False)
most_lethat
```

### Top 10 most deadly conflicts on the basis of total deaths

```python
deadly_conflict = df.groupby("conflict_name")["Total_deaths"].sum().sort_values(ascending = False).head(10)
deadly_conflict
```
