#### Q-12 calculate the average of sliding windows 3x3 

```python
rows, col = met.shape
output = np.zeros((rows-2, col-2))
for i in range(0, rows-2): 
    for j in range(0, col-2):
        windows = met[i: i+3, j:j+3]
        avg= windows.mean()
        output[i,j]= avg
print(output)
```

```python
maxout = np.zeros((rows-2, col-2))
for i in range(0, rows-2):
    for j in range(0, col-2):
        window= met[i: i+2, j:j+2]
        maxval = window.max()
        maxout[i,j] = maxval
print(maxout)
```

#### Q-14 Outlier 

```python
data = np.array([
    [45, 48, 50],
    [47, 49, 60],
    [46, 48, 47],
    [49, 50, 48]
])
```

```python
mean = data.mean()
std = data.std()
lower = mean - 2*std
upper = mean + 2*std
outlier = data[(data<lower) | (data> upper)]
print(outlier)
```
