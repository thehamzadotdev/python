## Array operations

### Arithmatic operations

```python
a1 = np.array([1,2,4,6,8,10])
a2 = np.array([1,3,5,7,9,11])
a1+a2
```

```python
a1-a2
```

```python
a1/a2
```

```python
a1*a2
```

```python
a1**a2
```

### Broadcasting

#### Broadcasting mean adding 1 number in all the elements of array

```python
arr = np.array([10,20,30,40,50,60])
arr +10
```

### shalow copy

```python
arr = np.arange(1,21)
```

```python
slice = arr[:5]
slice*10
```

```python
arr
```

#### See the example above the orignal array hasn't been changes while we have sliced the array and applied multiplication. This is shallow copy

## Metrics operations

```python
A = np.array([[2,3], [4,5]])
B= np.array([[6,7], [8,9]])
A
```

```python
A@B 
```

```python
np.dot(A,B)
```

```python
# @ or dot is used for multiplication of metrics
```

```python
A.T
```

```python
# T is use to take a transpose of metrics
```

### Stacking

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,7])
```

```python
np.hstack((a,b))
```

```python
np.vstack((a,b))
```

```python
np.column_stack((a,b))
```

### Split

```python
arr = np.arange(16).reshape(4,4)
arr
```

```python
np.vsplit(arr,4)
```

```python
np.hsplit(arr, 2)
```
