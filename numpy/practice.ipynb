```python
evenNumbers= arr%2 ==0
evenNumbers
```

```python
even_num_metrics = arr[evenNumbers]
even_num_metrics.reshape(3,5)
```

```python
oddNumbers= arr%2 ==1
oddNumbers
```

```python
odd_number_metrics = arr[oddNumbers]
odd_number_metrics.reshape(3,5)
```

```python
import matplotlib.pyplot as plt
even_matrix = even_num_metrics.reshape(3, 5)
odd_matrix = odd_number_metrics.reshape(3, 5)

plt.hist(even_matrix, bins=10)
plt.title("Even Numbers Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

plt.hist(odd_matrix, bins=10)
plt.title("Odd Numbers Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
