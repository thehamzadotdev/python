```python
import numpy as np 
```

```python
overs =20
ball_per_over= 6
total_balls = overs*ball_per_over
```

```python
possible_outcomes = np.array([0,1,2,3,4,5,6,-1])
probability = np.array([0.35, 0.30, 0.10, 0.02, 0.10, 0.05, 0.08, 0.10])
probability = probability / probability.sum()
```

```python
balls = np.random.choice(possible_outcomes, 
                         size = total_balls,
                         p= probability)
```

```python
total_runs = balls[balls>=0].sum()
total_wickets = np.sum(balls == -1)
total_runs
```

```python
over_wise = balls.reshape(overs, ball_per_over)
```

```python
over_wise_runs = np.array([
    over[over>=0].sum() for over in over_wise
])
```

```python
run_rate = total_runs/overs
```
